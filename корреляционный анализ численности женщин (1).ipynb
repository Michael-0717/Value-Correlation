```python
%matplotlib inline
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
data = np.array([[76.8, 2007], [76.8, 2008], [76.8, 2009], [76.8, 2010], [76.8, 2011], [76.9, 2012],[77.0, 2013], [77.1, 2014]] , dtype=float)
```

```python
data_xy = np.hsplit(data, (1,2))
```

```python
x = data_xy[0].flatten() # получим x-значение, численность женщин
y = data_xy[1].flatten() # получим y-значение, год
plt.scatter(x, y)
plt.xlabel("численность женщин")
plt.ylabel("год")
plt.show()
```

```python
pirson = st.pearsonr(x, y)[0]
if pirson > 0.7:
    print ("Пирсон: прямо пропорциональна численность женщин от года")
elif pirson < -0.7:
    print ("Пирсон: обратно пропорциональна численность женщин от года")
else:
    print ("Явной выраженной зависимости численности женщин от года нет")
```

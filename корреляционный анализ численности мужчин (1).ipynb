```python
%matplotlib inline
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
data = np.array([[66.0, 2007], [66.0, 2008], [65.9, 2009], [66.1, 2010], [66.1, 2011], [66.1, 2012],[66.3, 2013], [66.6, 2014]] , dtype=float)
```

```python
data_xy = np.hsplit(data, (1,2))
```

```python
x = data_xy[0].flatten() # получим x-значение, численность мужчин
y = data_xy[1].flatten() # получим y-значение, год
plt.scatter(x, y)
plt.xlabel("численность мужчин")
plt.ylabel("год")
plt.show()
```

```python
pirson = st.pearsonr(x, y)[0]
if pirson > 0.7:
    print ("Пирсон: прямо пропорциональна численность мужчин от года")
elif pirson < -0.7:
    print ("Пирсон: обратно пропорциональна численность мужчин от года")
else:
    print ("Явной выраженной зависимости численности мужчин от года нет")
```
